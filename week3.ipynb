- pandas apply()      .
-  cos  -1   .

```python
import pandas as pd

# -1   
def invert_value(value):
    return -value

x = np.linspace(0, 10, 100)
y = np.cos(x)

# y pandas DataFrame 
df = pd.DataFrame(y, columns=['y'])

# apply    y  
df['y_inverted'] = df['y'].apply(invert_value)

#  
plt.plot(x, df['y_inverted'])
plt.show()
```

##      .

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  #     y (sin )
y2 = np.cos(x)  #     y (cos )

# figure subplot 
fig, ax = plt.subplots(2, 1, figsize=(8, 6))  # 2 1 subplot, figure  8x6

ax[0].plot(x, y1) #    ( )
ax[1].plot(x, y2, 'r-', label='cos(x)') #    ( )

plt.tight_layout()
plt.show()
```
