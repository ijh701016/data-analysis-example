# 여러 그래프 한번에 표시하기
- sin함수와 cos함수를 그래프로 각각 그려 봅시다.
- 그 후 두 함수의 그래프를 한번에 출력하는 방법을 배워봅시다.

- 사인 함수는 numpy 라이브러리의 sin() 함수를 불러와 사용할 수 있습니다.
- 또한, linespace() 함수로 범위 내의 원하는 점의 개수를 생성할 수 있습니다.

```python
x = np.linspace(0, 10, 100)  # 0부터 10까지 100개의 점 생성
y = np.sin(x)  # y값을 sin 함수로 계산

plt.plot(x, y)
plt.show()
```

- 코사인 함수 또한 cos() 함수를 사용할 수 있습니다.

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)  # y값을 cos 함수로 계산

plt.plot(x, y)
plt.show()
```

- pandas의 apply() 함수로 데이터들의 값을 변경할 수 있습니다.
- 다음은 cos그래프의 데이터들에 -1을 곱하여 출력하는 코드입니다.

```python
import pandas as pd

# -1을 곱하는 함수 정의
def invert_value(value):
    return -value

x = np.linspace(0, 10, 100)
y = np.cos(x)

# y를 pandas DataFrame으로 변환
df = pd.DataFrame(y, columns=['y'])

# apply 함수를 사용해 반전된 y 값 생성
df['y_inverted'] = df['y'].apply(invert_value)

# 그래프 그리기
plt.plot(x, df['y_inverted'])
plt.show()
```

## 이제 두 함수로 만든 그래프를 합쳐봅시다.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # 첫 번째 그래프에 사용할 y값 (sin 함수)
y2 = np.cos(x)  # 두 번째 그래프에 사용할 y값 (cos 함수)

# figure와 subplot 생성
fig, ax = plt.subplots(2, 1, figsize=(8, 6))  # 2행 1열의 subplot, figure 크기는 8x6인치

ax[0].plot(x, y1) # 첫 번째 그래프 (사인 함수)
ax[1].plot(x, y2, 'r-', label='cos(x)') # 두 번째 그래프 (코사인 함수)

plt.tight_layout()
plt.show()
```

# 동물의숲 주민 데이터로 실습하기
- 항목별(성별, 종류, 성격 등)로 묶어 막대그래프 그려보기

```python
import pandas as pd
import koreanize_matplotlib # 그래프 내 한글 출력을 도와주는 라이브러리 (설치 필요)

villagers_df = pd.read_csv("villagers.csv") # csv파일 불러오기
villagers_df.head()
```

```python
# value_count() 함수로 개수를 샐 수 있습니다.
gender_counts = villagers_df['성별'].value_counts()

gender_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])

plt.title('주민 성별') # 그래프 제목
plt.xlabel('성별') # x축 제목
plt.ylabel('주민 수') # y축 제목
plt.xticks(rotation=0) # x축의 눈금 회전
plt.show()
```

- 아래에 주민의 종류는 원그래프로, 주민의 성격은 막대그래프(세로)로 그려봅시다.
